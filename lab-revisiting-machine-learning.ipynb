Use appropriate methods to clean the columns GEOCODE2, WEALTH1, ADI, DMA,and MSA.

```python
categorical['GEOCODE2'].value_counts(dropna=False)/len(categorical)
```

As the number of NaNs are just 0.3%, it is fast and safe to fill them with the mode, that is value 'A'

```python
categorical['GEOCODE2']=categorical['GEOCODE2'].fillna('A')
categorical['GEOCODE2'].value_counts(dropna=False)
```

```python
numerical['WEALTH1'].value_counts(dropna=False)/len(numerical)
```

As 47% of the values in WEALTH1 column are NaNs, it would be unsafe to fill them with mode or median value, as it would affect the analysis significantly. So it would be wise to label is as 'Other', after converting the column to Object.

```python
categorical['WEALTH1']=numerical['WEALTH1'].astype('object')
numerical = numerical.drop(['WEALTH1'], axis=1)
categorical['WEALTH1']=categorical['WEALTH1'].fillna('Other')
categorical['WEALTH1'].value_counts(dropna=False)
```

First, we'll check, how many of them are null and if they're very little, we'll replace them with the median value if the value of median and mean fall close to each other, for 2 reasons. First, the values in these columns are rather integer, even if they're represented as float, and having a mean value calculated will possibly turn out to be a float, and that will be a inconsistent value for the column as others are whole number. Therefore, it will be wise to replace it with a median value. Secondly, the median will preserve the normal distribution curve.

```python
for col in ['DMA','MSA','ADI']:
    display(pd.DataFrame(numerical[col].value_counts(dropna=False)).head())
```

```python
numerical[['ADI','MSA','DMA']].isna().sum()*100/len(numerical)
```

```python
numerical[['ADI','MSA','DMA']].describe().T
```

As we observe, the nulls are hardly just 0.1%, and the median-mean are nearly close to each other. We'll replace the NaNs with the median value.

```python
for col in ['ADI','MSA','DMA']:
    numerical[col]=numerical[col].fillna(numerical[col].median())
```
